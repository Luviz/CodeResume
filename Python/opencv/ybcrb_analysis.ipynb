```python
import cv2 as cv
from matplotlib import pyplot as plt

_ = plt.set_cmap("gray")
```

```python
# %matplotlib tk
img = cv.imread("./img/color-wheel-full.jpg")
yCrCb = cv.cvtColor(img, cv.COLOR_BGR2YCrCb)

fig, ax = plt.subplots(1,4)

fig.set_size_inches(20,6)

for ix, title in enumerate(["org", "y", "Chroma Red", "Chroma Blue"]):
    ax[ix].set_title(title)

ax[0].imshow(img[:,:,::-1])


ax[1].imshow(yCrCb[:,:,0])
ax[2].imshow(yCrCb[:,:,1])
ax[3].imshow(yCrCb[:,:,2])
```

```python
img = cv.imread("./img/color-wheel.jpg")
yCrCb = cv.cvtColor(img, cv.COLOR_BGR2YCrCb)

fig, ax = plt.subplots(1,4)

for ix, title in enumerate(["org", "y", "Chroma Red", "Chroma Blue"]):
    ax[ix].set_title(title)

fig.set_size_inches(20,6)

ax[0].imshow(img[:,:,::-1])

ax[1].imshow(yCrCb[:,:,0])
ax[2].imshow(yCrCb[:,:,1])
ax[3].imshow(yCrCb[:,:,2])
```

```python
y = yCrCb[:,:,0]
rgb_img = img[:,:,::-1]

fig, axies = plt.subplots(18, 3)
fig.set_size_inches(10, 15*3)
fig.set_tight_layout('tight')

y_value = 0
for ax_col in axies:
    mask = cv.inRange(y-y_value, 200, 255 + 10)
    ax_col[0].imshow(rgb_img)
    ax_col[1].imshow(mask)
    ax_col[2].imshow(cv.bitwise_and(rgb_img, rgb_img, mask=mask))
    ax_col[0].set_ylabel(f"{y_value}-{y_value+10}")
    y_value += 10
```
